# Lidando com dados nulos

```python
df.isnull()
```

```python
#Substitui os valores nulos por zero
df = df.fillna(0)
df
```

```python
#Veriicando que nao existe mais valores nulos
df.isnull().sum()
```

```python
#Obs:
#Remover os dados nulos: É possível remover as linhas ou colunas que possuem valores nulos utilizando o método dropna() .
#Esse método remove todas as linhas ou colunas que possuem pelo menos um valor nulo.

#Preencher os dados nulos: Utilizando o método fillna(), podemos preencher os valores nulos com um valor específico.
#Além disso, também é possível utilizar argumentos específicos do método fillna() como o method=”ffill” ou method=”bfill” 
#para preencher os valores nulos com o valor anterior ou posterior, respectivamente.

#Interpolar os dados nulos: É possível utilizar o método interpolate() para preencher os valores nulos com valores interpolados,
#ou seja, valores calculados a partir dos valores vizinhos.
```

# Removendo registros 

```python
#Registros com valores igual a zero
df.query('Valor == 0 | Condominio == 0').index
```

```python
registros_a_remover = df.query('Valor == 0 | Condominio == 0').index
```

```python
#Removendo registros com valores 0 ou condominio 0
df = df.drop(registros_a_remover, axis=0)
```

```python
df.query('Valor == 0 | Condominio == 0').index
```

```python
df.head()
```

```python
#Removendo a coluna que informa apartamento , ja que sabemos que todos os registros se refere a apartamento
#axis=1 remove coluna
#axis=0 remove linha
df = df.drop('Tipo', axis=1)

#É possivel remover linhas especificas
#df = df.drop([2,5,8], axis=0)
```

```python
df.head()
```

# Apartamento que possuem 1 quarto e aluguel menor que 1200


```python
#Selecionando apartamento que possuem apenas um quarto
selecao1 = df['Quartos'] == 1 
```

```python
df[selecao1]
```

```python
#Selecionando apenas apartamentos com alguem menor que 1200
selecao2 = df['Valor'] < 1200
```

```python
df[selecao2]
```

```python
#Juntando as selecoes
selecao_final = (selecao1) & (selecao2)
df[selecao_final]
```

```python
#Criando um dataframe com as selecoes
df_1 = df[selecao_final] 
df_1
```

# Apartamentos que possuem pelo menos 2 quartos, aluguel menor que 3000 e area maior que 70

```python
selecao = (df['Quartos'] >= 2) & ( df['Valor'] < 3000) & (df['Area'] > 70)
df[selecao]
```

```python
df_2 = df[selecao]
```

# Salvando os dados

```python
df.to_csv('dados_apartamentos.csv', index=False, sep=';')
df_1.to_csv('1 quarto aluguel menor 1200.csv', index=False, sep=';')
df_2.to_csv('2 ou mais quarto aluguel menor que 3000 area maior que 70', index=False , sep=';')
```
